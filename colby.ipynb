## Data Cleaning

```python
import numpy as np
import pandas as pd
import re
```

```python
ch89dg = pd.read_csv('./data/CH89DG.csv')
```

```python
ch89dg.head(2)
```

```python
def clean_header(df):
    '''fn should format all dataframes as described below'''
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace("-", "_").str.replace(" ", "").str.replace("__", "_")

    return df
```

```python
clean_header(ch89dg)
```

```python
ch89dg.replace(to_replace=' ', value='', regex=True, inplace=True)
ch89dg.replace(to_replace='^\.$', value='', regex=True, inplace=True)
```
